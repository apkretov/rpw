```python
import matplotlib.pyplot as plt

fx1 = (-4, -2)
fy1 = (4, 1)
fx2 = (-2, 1)
fy2 = (3, 2)
fx3 = (1, 4)
fy3 = (-1, 2)
colour = '#0099ff'

ax = plt.subplot() # show origin axis (x,y) in matplotlib plot @ https://stackoverflow.com/questions/25689238/show-origin-axis-x-y-in-matplotlib-plot
ax.plot(fx1, fy1, fx2, fy2, fx3, fy3, c=colour, linewidth=4)
ax.set_aspect('equal')
ax.grid(True)
ax.axhline(c='k') # 'k' black
ax.axvline(c='k')

last = len(fx1) - 1
ax.plot(fx1[last], fy1[last], 'o', c='w', markersize=10, markeredgecolor=colour)
ax.plot(fx3[0], fy3[0], 'o', c='w', markersize=10, markeredgecolor=colour)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def func(x):
    return 6 * x ** 2 + 5 * x - 1


colour = '#0099ff'
colour_dot = colour
x1 = -5
x2 = 5
x = -1

# x_array_1 = np.arange(x1, x2 + 1)
x_array = np.linspace(x1, x2, num=100)
y_array = func(x_array)

ax1 = plt.subplot()
ax1.plot(x_array, y_array, c=colour, linewidth=4)
ax1.grid(True)
ax1.axhline(c='k')
ax1.axvline(c='k')
y_dot = func(x)
ax1.plot(x, y_dot, 'o', c=colour_dot, markersize=10, markeredgecolor=colour)

plt.show()
print('x =', x, '\tf(x) =', y_dot)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math

def func(x):
    return math.log(x)


def func_array(x_array):
    result = []
    for item in x_array:
        result.append(func(item))
    return result


colour = '#0099ff'
colour_dot = colour
dot_qty = 100
x1 = 0.1
x2 = 5
x = math.e

x_array = np.linspace(x1, x2, num=dot_qty)
y_array = func_array(x_array)

ax1 = plt.subplot()
ax1.plot(x_array, y_array, c=colour, linewidth=4)
ax1.grid(True)
ax1.axhline(c='k')
ax1.axvline(c='k')
y_dot = func(x)
ax1.plot(x, y_dot, 'o', c=colour_dot, markersize=10, markeredgecolor=colour)

print('x =', x, '\tf(x) =', y_dot)
plt.show()
```
