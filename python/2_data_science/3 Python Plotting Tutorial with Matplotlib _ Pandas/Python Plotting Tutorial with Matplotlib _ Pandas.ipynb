### Line Graph 03:49

```python
gas = pd.read_csv('gas_prices.csv')
gas
```

```python
plt.figure(figsize=(8, 5))
plt.title('Gas Prices over Time, USD')

plt.plot(gas.Year, gas.USA, 'b.-', label='USA')
plt.plot(gas.Year, gas.Canada, 'r.-', label='Canada')
plt.plot(gas.Year, gas['South Korea'], 'g.-', label='South Korea')
plt.plot(gas.Year, gas.Australia, 'y.-', label='Australia')

plt.xticks(gas.Year[::2]) #Years step = 2.

plt.xlabel('Year')
plt.ylabel('USD')

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.title('Gas Prices over Time, USD')

for country in gas: #All the countries in a loop.
    if country != 'Year':
        plt.plot(gas.Year, gas[country], marker='.')

plt.xticks(gas.Year[::2])

plt.xlabel('Year')
plt.ylabel('USD')

# plt.legend() #It does not work by some reason...
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.title('Gas Prices over Time, USD', fontdict={'fontweight': 'bold', 'fontsize': 18}) #The font in dictionary parameter.

counries_to_look_at = ['Australia', 'USA', 'Canada', 'South Korea'] #Selected countries in a loop.
for country in counries_to_look_at:
    if country != 'Year':
        plt.plot(gas.Year, gas[country], marker='.')

plt.xticks(gas.Year[::2].tolist() + [2010]) #Extend the years.
plt.xlabel('Year')
plt.ylabel('USD')

plt.savefig('Gas_price_figure.png', dpi=300)
plt.show()
```

### Box & Wiskers Chart 47:49

```python
plt.style.use('default')

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']

labels = ('FC Barcelona', 'Real Madrid')

plt.boxplot([barcelona, madrid], labels=labels)
plt.show()
```

```python
plt.style.use('default')
plt.figure(figsize=(5, 8))

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

labels = ('FC Barcelona', 'Real Madrid', 'New England Revolution')

plt.boxplot([barcelona, madrid, revs], labels=labels)

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')
plt.show()
```

```python
plt.style.use('default')
plt.figure(figsize=(5, 8))

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

labels = ('FC Barcelona', 'Real Madrid', 'New England Revolution')

boxes = plt.boxplot([barcelona, madrid, revs], labels=labels)

for box in boxes['boxes']:
    box.set(color='#4286f4', linewidth=2) #Set edge colour.

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')
plt.show()
```

```python
plt.style.use('default')
plt.figure(figsize=(5, 8))

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

labels = ('FC Barcelona', 'Real Madrid', 'New England Revolution')

boxes = plt.boxplot([barcelona, madrid, revs], labels=labels, patch_artist=True)

for box in boxes['boxes']:
    box.set(color='#4286f4', linewidth=2)
    box.set(facecolor='#e0e0e0') #Change Fill colour.

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')
plt.show()
```

```python
plt.style.use('default')
plt.figure(figsize=(5, 8))

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

labels = ('FC Barcelona', 'Real Madrid', 'New England Revolution')

boxes = plt.boxplot([barcelona, madrid, revs], labels=labels, patch_artist=True, 
                    medianprops={'linewidth': 2}) #Median line width.

for box in boxes['boxes']:
    box.set(color='#4286f4', linewidth=2)
    box.set(facecolor='#e0e0e0')

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')
plt.show()
```
