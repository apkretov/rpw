```python
import pandas as pd

df = pd.read_csv('pokemon_data.csv')
df.loc[df['Type 1'] == 'Grass'].head(3)
```

```python
df_modified = df.loc[df['Type 1'] == 'Fire']
df_modified.loc[df_modified['Type 1'] == 'Fire', 'Type 1'] = 'Flamer' #Modify column Type 1
df_modified.head(3)
```

```python
df_modified = df.loc[df['Type 1'] == 'Fire']
df_modified.loc[df_modified['Type 1'] == 'Fire', 'Legendary'] = True #Modify another column
df_modified.head(3)
```

```python
df_modified = df.loc[df['Type 1'] == 'Fire']
df_modified.loc[df_modified.Speed > 100, ['Generation', 'Legendary']] = 'TEST VALUE' #Modify multiple columns equally.
df_modified[df_modified.Generation == 'TEST VALUE'].head(3)
```

```python
df_modified = df.loc[df['Type 1'] == 'Fire']
df_modified.loc[df_modified.Speed > 100, ['Generation', 'Legendary']] = ['Test 1', 'Test 2'] #Modify multiple columns differently.
df_modified[(df_modified.Generation == 'Test 1') | (df_modified.Legendary == 'Test 2')].head(3)
```
