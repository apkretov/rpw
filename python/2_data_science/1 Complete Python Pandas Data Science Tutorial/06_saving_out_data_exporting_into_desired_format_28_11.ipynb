```python
import pandas as pd
df = pd.read_csv('pokemon_data.csv')
df['Total'] = df.iloc[:, 4:10].sum(axis=1)

cols = list(df.columns)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]

df.to_csv('modified.csv')

df_modified = pd.read_csv('modified.csv')
df_modified.head(3)
```

```python
df.to_csv('modified_no_index.csv', index=False)

df_modified_no_index = pd.read_csv('modified_no_index.csv')
df_modified_no_index.head(3)
```

```python
df.to_excel('modified.xlsx', index=False)

df_excel = pd.read_excel('modified.xlsx')
df_excel.head(3)
```

```python
df.to_csv('modified.txt', index=False, sep='\t')

df_txt = pd.read_csv('modified.txt')
df_txt.head(3)
```
