```python
import pandas as pd

df = pd.read_csv('pokemon_data.csv')
df.loc[df['Type 1'] == 'Grass'].head(3)
```

```python
df.loc[df.Name.str.contains('Mega')].head(3)
```

```python
df.loc[~df.Name.str.contains('Mega')].head(3)
```

```python
import re #regex

df.loc[df['Type 1'].str.contains('Fire|Grass', regex=True)].head(5)
```

```python
import re #regex

df.loc[df['Type 1'].str.contains('fire|grass', flags=re.I, regex=True)].head(5) #re.I - ignore case
```

```python
df.loc[df.Name.str.contains('pi[a-z]*', flags=re.I, regex=True)].head(3) #Name contains pi
```

```python
df.loc[df.Name.str.contains('^pi[a-z]*', flags=re.I, regex=True)].head(3) #Name does not contain pi
```
