### Line Number Two 17:45

```python
plt.plot(x, y, 'b^--', label='2x')

x2 = np.arange(0, 4.5, 0.5) #Return evenly spaced values within a given interval. @ https://docs.scipy.org/doc/numpy/reference/generated/numpy.arange.html
plt.plot(x2, x2 ** 2, 'r', label='x^2') #Line Number Two

plt.title('Our First Graph', fontdict={'fontname': 'MathJax_Math', 'fontsize': 20})
plt.xlabel('X Axis', fontdict={'fontname': 'MathJax_Math'})
plt.ylabel('Y Axis', fontdict={'fontname': 'MathJax_Math'})

plt.xticks([0, 1, 2, 3, 4])
plt.legend()
plt.show()
print(f'x2: {x2}')
```

```python
plt.plot(x, y, 'b^-', label='2x')

plt.plot(x2[:6], x2[:6] ** 2, 'r', label='x^2')
plt.plot(x2[5:], x2[5:] ** 2, 'r--') #A dashed-line projection.

plt.title('Our First Graph', fontdict={'fontname': 'MathJax_Math', 'fontsize': 20})
plt.xlabel('X Axis', fontdict={'fontname': 'MathJax_Math'})
plt.ylabel('Y Axis', fontdict={'fontname': 'MathJax_Math'})

plt.xticks([0, 1, 2, 3, 4])
plt.legend()
plt.show()
print(f'x : {x}')
print(f'x2: {x2}')
```

### Bar Chart 27:20

```python
labels = ['A', 'B', 'C']
values = [1, 4, 2]

plt.bar(labels, values)
plt.show()
```

```python
bars = plt.bar(labels, values)

bars[0].set_hatch('/') #Set the hatch. E.g: Hatch Demo @ https://matplotlib.org/3.1.1/gallery/shapes_and_collections/hatch_demo.html
bars[1].set_hatch('o')
bars[2].set_hatch('*')

plt.show()
```

```python
bars = plt.bar(labels, values)

patterns = ['/', 'o', '*'] #Set the hatches in a loop.
for bar in bars:
    bar.set_hatch(patterns.pop())

plt.show()
```
