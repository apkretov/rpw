## 4. Python Matplotlib Tutorial – Pyplot

### a. plot()

```python
import matplotlib.pyplot as plt

plt.plot([2,3,4,5])
plt.xlabel('Actual birth weight')
plt.ylabel('Estimated birth weight')
plt.show()
```

```python
plt.plot([2,3,4,5],[3,8,10,12])
plt.xlabel('Actual birth weight')
plt.ylabel('Estimated birth weight')
plt.show()
```

### b. Formatting your Python Plot

```python
plt.plot([2,3,4,5],[3,8,10,12],'s') #'s' square markers.
plt.show()
```

```python
plt.plot([2,3,4,5],[3,8,10,12],'gs') #'g' green 's' square markers.
plt.show()
```

```python
plt.plot([2,3,4,5],[3,8,10,12],'gs')
plt.axis([0,7,0,21]) #X & Y axises
plt.show()
```

```python
import numpy as np
t=np.arange(0,5,0.2)
plt.plot(t,t,'r--', t,t**3,'b^', t,t**2,'gs') ##Three graphs on one chart.
plt.show()
```

## 5. Python Matplotlib Keyword Strings

```python
data = {'a': np.arange(50),
      'c': np.random.randint(0, 50, 50),
      'd': np.random.randn(50)}

data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100
plt.scatter('a', 'b', c='c', s='d', data=data)
plt.show()
```

```python
n = 100 #More dots.
data = {'a': np.arange(n),
      'c': np.random.randint(0, n, n),
      'd': np.random.randn(n)}

data['b'] = data['a'] + 10 * np.random.randn(n)
data['d'] = np.abs(data['d']) * 300 #Bigger size.

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.show()
```

## 6. Categorical Variables to Python Plotting

```python
# We can pass categorical variables to Python plotting functions.
names = ["Dingos", "Wild Cats", "Tigers"]
values = [1, 11, 111]
plt.figure(1, figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.show()
```

```python
plt.figure(1, figsize=(9, 3))

plt.bar(names, values) #No plt.subplot(131)
plt.show()
```

```python
plt.figure(1, figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)

plt.subplot(132)
plt.scatter(names, values)

plt.show()
```

```python
plt.figure(1, figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)

plt.subplot(132)
plt.scatter(names, values)

plt.subplot(133)
plt.plot(names,values)

plt.suptitle('Varsity')
plt.show()
```

## 7. Some Line Properties of Matplotlib

### a. Linewidth

```python
plt.plot([1,2,3], [2,4,9], linewidth=4.0)
plt.show()
```

### b. Alpha

```python
# We can use the alpha channel to create prettier plots by softening colors.
plt.plot([1,2,3], [2,4,9], alpha=5.5)
plt.show()
```

```python
plt.plot([1,2,3], [2,4,9]) #No alpha that stands for translucency.
plt.show()
```

### c. Antialiased

```python
plt.plot([1,2,3],[2,4,9],antialiased=True)
plt.show()
```

```python
plt.plot([1,2,3],[2,4,9],antialiased=True, linewidth=10)
plt.show()
```

```python
plt.plot([1,2,3],[2,4,9], linewidth=10)
plt.show()
```

### d. Color or c

```python
plt.plot([1,2,3],[2,4,9],color='Chartreuse')
plt.show()
```

### e. Dashes

```python
plt.plot([1,2,3], [2,4,9], dashes=[1,2, 4,4])
plt.show()
```

```python
#set_dashes(self, seq) @ https://matplotlib.org/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D.set_dashes
plt.plot([1,2,3], [2,4,9], dashes=(5,2, 1,2)) #Describes a sequence of 5 point and 1 point dashes separated by 2 point spaces.
plt.show()
```

### f. Linestyle or ls

```python
plt.plot([1,2,3],[2,4,9],linestyle='steps')
plt.show()
```

```python
plt.plot([1,2,3],[2,4,9],linestyle=':') #set_linestyle(self, ls)[source] @ https://matplotlib.org/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D.set_linestyle
plt.show()
```

```python
plt.plot([1,2,3], [2,4,9], linestyle='steps-mid')  # set_drawstyle(self, drawstyle) @ https://matplotlib.org/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D.set_drawstyle
plt.show()
```

```python
plt.plot([1,2,3], [2,4,9], linestyle='steps-post')  # set_drawstyle(self, drawstyle) @ https://matplotlib.org/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D.set_drawstyle
plt.show()
```

### g. Marker

```python
plt.plot([1,2,3],[2,4,9],marker='+')
plt.show()
```

### h. Markeredgecolor

```python
plt.plot([1,2,3],[2,4,9],marker='+',markeredgecolor='brown')
plt.show()
```

### i. Markeredgewidth

```python
# This lets us decide how thick we want our markers to be.
plt.plot([1,2,3],[2,4,9],marker='+',markeredgewidth=0.4)
plt.show()
```

### j. Markerfacecolor and Markersize

```python
plt.plot([1,2,3],[2,4,9],marker='.',markerfacecolor='orange',markersize=13.0)
plt.show()
```

### k. Markevery

```python
# This parameter lets us decide at what parameters we want to put markers.
plt.plot([1,2,3],[2,4,9],marker='.',markerfacecolor='orange',markersize=13.0,markevery=2)
plt.show()
```

```python
plt.plot([1,2,3],[2,4,9],marker='.',markerfacecolor='orange',markersize=13.0,markevery=1)
plt.show()
```

### l. Zorder

```python
# This lets us decide which plots will show up in the front and which to send to back.
plt.plot([1,2,3],[2,4,9],zorder=1,linewidth=4)
plt.plot([1,2,6,9],[2,4,9,10],zorder=2,linewidth=4)
plt.show()
```

## 8. Showing a Grid in Python Plot
